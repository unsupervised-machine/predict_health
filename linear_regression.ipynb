```python
# load packages
import pandas as pd
import sklearn
from matplotlib import pyplot

pd.set_option('display.max_columns', None)

```

```python
# load data
health_data = pd.read_csv('./data/data01.csv')
health_data
```

```python
# change data type categorical columns
cols = ['group', 'ID', 'outcome', 'gendera', 'hypertensive',
        'atrialfibrillation', 'CHD with no MI', 'diabetes', 'deficiencyanemias',
        'depression', 'Hyperlipemia', 'Renal failure', 'COPD']

health_data[cols] = health_data[cols].astype('category')
```

```python
# drop rows missing outcome variable
health_data = health_data.dropna(subset=['outcome'])
```

```python
# impute numeric columns with gender mean
# test = health_data.copy()
# numeric_cols = test.select_dtypes(include=['int64', 'float64']).columns
# cat_cols = test.select_dtypes(include=['category']).columns

# for i in numeric_cols:
#     test[i] = test[i].fillna(test.groupby('gendera')[i].transform('mean'))

# test.isnull().sum()
```

```python
# impute numeric columns with gender mean
numeric_cols = health_data.select_dtypes(include=['int64', 'float64']).columns
# cat_cols = health_data.select_dtypes(include=['category']).columns

for i in numeric_cols:
    health_data[i] = health_data[i].fillna(health_data.groupby('gendera')[i].transform('mean'))

# health_data.info()
```
